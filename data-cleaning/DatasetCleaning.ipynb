```python
import random
import pandas as pd
# del df_rating
df_rating = pd.read_csv('/content/drive/MyDrive/Movie Datasets/Reviews and Ratings/{}.csv'.format(1))
df_extracted = pd.DataFrame(columns=df_rating.columns)
for i in range(1, 6):
  df_rating = pd.read_csv('/content/drive/MyDrive/Movie Datasets/Reviews and Ratings/{}.csv'.format(i))
  df_rating['Frequency'] = df_rating.groupby('movie')['movie'].transform('count')
  df_rating.sort_values('Frequency', inplace=True, ascending=False)
  # df_rating.drop('Frequency', axis=1)
  set_of_movie = set()
  count = 0
  for index, row in df_rating.iterrows():
    if row['movie'] in set_of_movie:
      continue
    df_selected = df_rating.loc[df_rating['movie'] == row['movie']]
    df_selected = df_selected.sample(10)
    df_extracted = df_extracted.append(df_selected)
    df_extracted = df_extracted.drop('Frequency', axis=1)
    set_of_movie.add(row['movie'])
    count = count + 1
    if count == 2000:
      break
    del df_selected
  del df_rating
```
